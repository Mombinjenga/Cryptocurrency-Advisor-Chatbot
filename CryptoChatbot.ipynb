```python
# Part 1 — imports, load dataset, quick sanity checks

# Libraries
import pandas as pd
import numpy as np

# Filename
FILE = "crypto_data.csv"

# Load and parse dates
df = pd.read_csv(FILE, parse_dates=["date"], infer_datetime_format=True)

# Sort by date and reset index
df = df.sort_values("date").reset_index(drop=True)

# Quick checks
print("rows, cols:", df.shape)
print("date range:", df['date'].min(), "to", df['date'].max())
print("missing per column:\n", df.isna().sum())

# Show first 5 rows
df.head()
```

```python
# Part 2 — clean dataset and create prediction targets

# Remove any rows with missing values
df = df.dropna()

# Create next-day closing price target
df['target_close'] = df['close'].shift(-1)

# Create direction target: 1 = up, 0 = down
df['target_direction'] = (df['target_close'] > df['close']).astype(int)

# Drop last row (because the target becomes NaN)
df = df[:-1]

# Quick check
df[['close', 'target_close', 'target_direction']].head()
```
