```python
import pandas as pd

# Load the full dataset
df = pd.read_csv('crypto_data.csv')

print(f"Original shape: {df.shape}")

# FILTER TO BITCOIN ONLY
btc_df = df[df['symbol'] == 'BTC'].copy()

# Or alternatively: df[df['name'] == 'Bitcoin'] or df[df['slug'] == 'bitcoin']

print(f"Bitcoin rows only: {btc_df.shape[0]} rows")

# Clean and prepare
btc_df['date'] = pd.to_datetime(btc_df['date'])
btc_df = btc_df.sort_values('date').reset_index(drop=True)

# Drop unnecessary metadata columns
btc_df = btc_df.drop(['slug', 'symbol', 'name', 'ranknow'], axis=1)

# Rename market to market_cap
btc_df = btc_df.rename(columns={'market': 'market_cap'})

# Set date as index (great for time-series modeling)
btc_df = btc_df.set_index('date')

print("\nBitcoin dataset ready for modeling!")
print(f"Final shape: {btc_df.shape}")
print(f"Date range: {btc_df.index.min().date()} to {btc_df.index.max().date()}")
print(f"Columns: {btc_df.columns.tolist()}")

btc_df.head()
```
